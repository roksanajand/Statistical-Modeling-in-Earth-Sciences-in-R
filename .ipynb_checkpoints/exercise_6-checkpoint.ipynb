## Splitting Data into Training and Testing Sets

### Data Preparation and Cleaning for Weather Dataset
This task involves preparing and filtering a weather dataset for further analysis. The dataset contains 22 variables describing meteorological stations in Germany. The following steps are included:

1. **Data Loading and Initial Overview**: The dataset is loaded, and the data structure and summary statistics are reviewed to understand its content.
2. **Feature Selection**: Irrelevant variables such as sDWD_ID,id, station name, federal state, period are excluded, as they do not contribute to predicting average annual rainfall.
3. **Missing Data Handling**: Rows with missing values are removed, and the number of rows deleted due to missing data is calculated.
4. **Dataset Dimensions**: The dimensions of the cleaned dataset (after removing irrelevant variables and rows with missing values) are displayed to confirm its readiness for analysis.

```r

library(caTools)
library(GGally)
library(corrplot)
```

```r
dane <- read.csv2("weather.csv")
head(dane)

str(dane)
summary(dane)
```

```r
colnames(dane)
dane_nowe = dane[c( 'LAT', 'LON', 'ALTITUDE', 'RECORD.LENGTH','MEAN.ANNUAL.AIR.TEMP', 'MEAN.MONTHLY.MAX.TEMP', 'MEAN.MONTHLY.MIN.TEMP', 'MEAN.ANNUAL.WIND.SPEED', 'MEAN.CLOUD.COVER','MEAN.ANNUAL.SUNSHINE', 'MEAN.ANNUAL.RAINFALL', 'MAX.MONTHLY.WIND.SPEED', 'MAX.AIR.TEMP', 'MAX.WIND.SPEED', 'MAX.RAINFALL', 'MIN.AIR.TEMP', 'MEAN.RANGE.AIR.TEMP')]
```

```r
sapply(dane_nowe, function(x) sum(x == ""))
dane_nowe[dane_nowe == ""] <- NA
liczba_wierszy_przed <- nrow(dane_nowe)
dane_czyste <- na.omit(dane_nowe)
liczba_wierszy_po <- nrow(dane_czyste)
liczba_usunietych_wierszy <- liczba_wierszy_przed - liczba_wierszy_po
liczba_usunietych_wierszy
```

```r

dim(dane_czyste)
```

### Splitting the Dataset into Training and Testing Sets



This step involves dividing the cleaned weather dataset into two subsets: a training set (70%) and a testing set (30%). The sample.split() function from the caTools library is used for the split, ensuring proper distribution. The training set will be used for building predictive models, while the testing set will evaluate the model's performance.

```r
split = sample.split(dane_czyste, SplitRatio = 0.3 , group = NULL)
dane_uczace = subset(dane_czyste, split == FALSE)
head(dane_uczace)

```

```r
dane_testowe = subset(dane_czyste, split == TRUE)
head(dane_testowe)
```

This step isolates the target variable, MEAN.ANNUAL.RAINFALL, as a vector for both training and testing datasets. It also creates matrices of independent variables for model building. Pairwise scatter plots are generated using the ggpairs() function from the GGally package to visualize relationships between variables. Additionally, a correlation matrix is computed and displayed using the corrplot() function to analyze the strength and direction of relationships between variables

```r
vec_train = as.vector(dane_uczace['MEAN.ANNUAL.RAINFALL'])
vec_train
vec_test = as.vector(dane_testowe['MEAN.ANNUAL.RAINFALL'])
vec_test
```

```r
matrix_test = data.matrix(dane_testowe[colnames(dane_testowe) != 'MEAN.ANNUAL.RAINFALL']) 
head(matrix_test)
matrix_train = data.matrix(dane_uczace[colnames(dane_uczace) != 'MEAN.ANNUAL.RAINFALL'])
head(matrix_train)

```

```r
ggpairs(dane_uczace)
```

```r
korelacja <-  cor(dane_czyste)
head(korelacja)
corrplot(korelacja, method = "number")
```

This step filters pairs of variables with a correlation coefficient |r| ≥ 0.5 relative to the target variable MEAN.ANNUAL.RAINFALL. Only highly correlated variables are selected for further analysis. A pairwise scatter plot matrix is generated using ggpairs() to visualize relationships among the selected variables, aiding in identifying meaningful predictors for the model.

```r
filter = abs(cor(dane_czyste['MEAN.ANNUAL.RAINFALL'],dane_czyste)) >= 0.5
filter
```

```r
ggpairs(dane_czyste[as.vector(filter)])
```
