# Polynomial Regression

### Linear Regression Model for Geophysical Parameters

This task involves reading geophysical parameter data from a file, visualizing the data using a scatter plot.

```r
my_data <- read.table("dane.txt", header = TRUE)

library(ggplot2)
ggplot(my_data,aes(X,Y))+
  geom_point(color="blue", pch =23, fill = "blue")
```

### Linear Regression Model: Evaluation and Analysis

```r
model<-lm(Y~X, data=my_data) #model regresji liniowej

summary(model) #statystyki modelu

predictions <- predict(model, data=my_data)

mse <- mean((predictions - my_data$Y)^2)

rmse <- sqrt(mse)
rmse2 <- sqrt(mean(residuals(model)^2))
cat("Model Evaluation Metrics:\n")
cat("Mean Squared Error (MSE):", round(mse, 4), "\n")
cat("Root Mean Squared Error (RMSE):", round(rmse, 4), "\n")
cat("Root Mean Squared Error (RMSE) using residuals:", round(rmse2, 4), "\n")
```

```r
coefficients(model)  
intercept <- coef(model)[1]  
slope <- coef(model)[2]      
cat("Estimated Model Parameters:\n")
cat("Intercept:", round(intercept, 4), "\n")
cat("Slope (coefficient for X):", round(slope, 4), "\n")
```

```r

r_squared <- summary(model)$r.squared
r_squared
#he R² value ranges from 0 to 1.
#The higher the R² value, the better the model explains the variability of the data.
#A value of 1 indicates that the model perfectly fits the data.
```

```r
ggplot(my_data,aes(X,Y))+
  geom_point(color="blue") + 
  geom_smooth(method="lm", se=FALSE, color="red") 

```

This task involves fitting polynomial regression models of degrees 2 to 10 to the given dataset. The polynomial curves are plotted alongside the measured data points to visually assess the fit of each model.

```r
p <- ggplot(my_data, aes(x=X, y=Y)) +
  geom_point(color="blue", alpha=0.5) 

for (i in 2:10) {
  p <- p + 
    geom_smooth(method="lm", formula=y ~ poly(x, i), se=FALSE, color=rainbow(9)[i-1], linewidth=1)
  print(p)
}

print(p)

```

This task computes regression coefficients, residual sum of squares (RSS), and the coefficient of determination (R²) for polynomial regression models of degrees 2 to 10. The results are summarized in a table to compare the performance of each model.

```r
results <- data.frame(Degree = integer(), Coefficients = list(), RSS = numeric(), R_squared = numeric())

for (i in 2:10) {

  model <- lm(Y ~ poly(X, i,raw=TRUE), data=my_data)

  coefficients <- coef(model)
  
  residuals <- resid(model)
  rss <- sum(residuals^2)
  
  r_squared <- summary(model)$r.squared
  
  results <- rbind(results, data.frame(Degree = i, Coefficients = I(list(coefficients)), RSS = rss, R_squared = r_squared))
}

print(results)
```
